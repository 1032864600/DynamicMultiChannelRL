```python
import pandas as pd
import numpy as np 
```

```python
len_desired = 5500
rand_1 = [np.random.randint(2) for i in range(len_desired)]
rand_2 = [np.random.randint(2) for i in range(len_desired)]
rand_3 = [np.random.randint(2) for i in range(len_desired)]
```

```python
def gen_list(in_com):
    if(in_com==0):
        return rand_1
    if(in_com==1):
        return rand_2
    if(in_com==2):
        return rand_3
    if(in_com==3):
        return [abs(rand_1[i]-1) for i in range(len_desired)]
    if(in_com==4):
        return [abs(rand_2[i]-1) for i in range(len_desired)]
    if(in_com==5):
        return [abs(rand_3[i]-1) for i in range(len_desired)]
```

```python
data_perf = pd.DataFrame()

data_perf["channel0"] = rand_1
data_perf["channel1"] = rand_2
data_perf["channel2"] = rand_3


for i in range(13):
    choice = np.random.randint(6)
    data_perf["channel"+str(i+4)] = gen_list(choice)
```
